```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.preprocessing import StandardScaler

# 加载数据集
medal_counts = pd.read_csv('D:/C/summerOly_medal_counts.csv', encoding='ISO-8859-1')
programs = pd.read_csv('D:/C/summerOly_programs.csv', encoding='ISO-8859-1')
athletes = pd.read_csv('D:/C/summerOly_athletes.csv', encoding='ISO-8859-1')
hosts = pd.read_csv('D:/C/summerOly_hosts.csv', encoding='ISO-8859-1')
team_size = pd.read_csv('D:/C/Per Size Of NOC.csv')
first_participation = pd.read_csv('D:/C/First time of NOC.csv')

# 数据清洗和预处理
hosts.columns = hosts.columns.str.strip()
hosts.rename(columns={'Host Country': 'Host'}, inplace=True)
medal_counts['Year'] = pd.to_numeric(medal_counts['Year'], errors='coerce')
programs_years = programs.loc[:, programs.columns.str.isnumeric()]
medal_counts = medal_counts.merge(hosts, left_on='Year', right_on='Year', how='left')
medal_counts['is_host'] = (medal_counts['NOC'] == medal_counts['Host']).astype(int)
features = ['Year', 'Gold', 'Silver', 'Bronze', 'Total', 'is_host', 'NOC']
data = medal_counts[features]
data = data.dropna()

# 合并队伍规模数据
data = data.merge(team_size, on=['NOC', 'Year'], how='left')

# 合并首次参赛年份数据
data = data.merge(first_participation, on='NOC', how='left')

# 处理缺失值
data['First Participation Year'] = data['First Participation Year'].fillna(0)
data['Team Size'] = data['Team Size'].fillna(0)

# 创建新特征：当前年份与首次参赛年份的差值
data['Years Since First Participation'] = data['Year'] - data['First Participation Year']

# 添加历史成绩特征
data["Gold_h"] = data.groupby('NOC')['Gold'].shift(1).fillna(0)
data["Total_h"] = data.groupby('NOC')['Total'].shift(1).fillna(0)

# 更新特征矩阵，去除 gdp 和 population 特征，添加 Team Size 特征
X = data[['Year', 'is_host', 'Gold_h', 'Total_h', 'Team Size', 'Years Since First Participation']]
y_gold = data['Gold']

# 随机森林回归与超参数调优
# 划分训练集和测试集
X_train, X_test, y_train_gold, y_test_gold = train_test_split(X, y_gold, test_size=0.2, random_state=42)

# 超参数调优
param_grid = {
    'n_estimators': [100, 200, 300],
    'max_depth': [None, 10, 20],
    'min_samples_split': [1, 2, 5, 10],
    'min_samples_leaf': [1, 2, 4]
}
rf_model = RandomForestRegressor(random_state=42)
grid_search = GridSearchCV(estimator=rf_model, param_grid=param_grid, scoring='neg_mean_squared_error', verbose=2, n_jobs=-1)
grid_search.fit(X_train, y_train_gold)

# 最佳参数
best_rf_model = grid_search.best_estimator_
print("Best Parameters:", grid_search.best_params_)

# 评估优化后的模型
gold_pred_rf = best_rf_model.predict(X_test)
gold_mse_rf = mean_squared_error(y_test_gold, gold_pred_rf)
gold_r2_rf = r2_score(y_test_gold, gold_pred_rf)
print("Gold Medal Optimized Random Forest Model MSE:", gold_mse_rf)
print("Gold Medal Optimized Random Forest Model R2 Score:", gold_r2_rf)

# 特征重要性
feature_importances = pd.Series(best_rf_model.feature_importances_, index=X.columns)
feature_importances.sort_values().plot(kind='barh', title='Feature Importance (Optimized Random Forest)', color='blue')
plt.xlabel('Importance')
plt.ylabel('Features')
plt.show()

# 可视化模型结果
# 绘制特征重要性图
plt.figure(figsize=(10, 6))
feature_importances.sort_values().plot(kind='barh', title='Feature Importance (Optimized Random Forest)', color='red')
plt.xlabel('Importance')
plt.ylabel('Features')
plt.show()

# 绘制实际值与预测值散点图
plt.figure(figsize=(10, 6))
sns.scatterplot(x=y_test_gold, y=gold_pred_rf, color='yellow', edgecolor='b', s=80)
plt.plot([y_test_gold.min(), y_test_gold.max()], [y_test_gold.min(), y_test_gold.max()], '-r', linewidth=2)
plt.title('Actual vs Predicted Gold Medals')
plt.xlabel('Actual Gold Medals')
plt.ylabel('Predicted Gold Medals')
plt.show()

# 绘制残差图
residuals = y_test_gold - gold_pred_rf
plt.figure(figsize=(10, 6))
sns.histplot(residuals, kde=True, color='pink', bins=20)
plt.xlabel('Residuals')
plt.title('Residuals Distribution')
plt.ylabel('Frequency')
plt.show()

# 预测2028年金牌数
example_2028 = pd.DataFrame({
    'Year': [2028],
    'is_host': [1],
    'Total_h': [120],
    'Gold_h': [40],
    'Team Size': [200],  # 假设队伍规模为 200
    'Years Since First Participation': [2028 - 1900]  # 假设该国家首次参赛年份为1900
})
gold_pred_2028 = best_rf_model.predict(example_2028)
print("Predicted Gold Medals for 2028:", gold_pred_2028[0])

```
